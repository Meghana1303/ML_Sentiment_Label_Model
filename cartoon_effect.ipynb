```
image = cv2.imread('deepveer.jpg')
```

```
plt.imshow(image)
plt.show()
```

```
#blurred image
blurred_img=cv2.medianBlur(image,15)
plt.imshow(blurred_img)
plt.show()
```

```
#detect edges on blurred image
blur_edges=cv2.Canny(blurred_img,75,150)
plt.imshow(blur_edges)
plt.show()
```

```
#edges
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray=cv2.medianBlur(gray,5)
edges=cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,15,9)
plt.imshow(edges)
plt.show()
```

```
#color
color =cv2.bilateralFilter(image,15,500,500)
plt.imshow(color)
plt.show()
```

```
#cartoon
cartoon=cv2.bitwise_and(color,color,mask=edges)
plt.imshow(cartoon)
plt.show()
```
